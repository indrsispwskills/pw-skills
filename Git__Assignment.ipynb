Q 4. Tasks to Be Performed:

1. Put master.txt on master branch, stage and commit

2. Create 3 branches: public 1, public 2 and private

3. Put public1.txt on public 1 branch, stage and commit

4. Merge public 1 on master branch

5. Merge public 2 on master branch

6. Edit master.txt on private branch, stage and commit

7. Now update branch public 1 and public 2 with new master code in private

8. Also update new master code on master

9. Finally update all the code on the private branch

**#### Initialize repo**

mkdir GitAssignment4
cd GitAssignment4
git init

touch master.txt
git add master.txt
git commit -m "Added master.txt"

**#### 2. Create branches**

git branch public1
git branch public2
git branch private

**#### 3. public1 branch**

git checkout public1
touch public1.txt
git add public1.txt
git commit -m "Added public1.txt"

**#### 4. Merge public1 into master**

git checkout master
git merge public1

**#### 5. Merge public2 into master**

git merge public2

**#### 6. Edit master.txt in private**

git checkout private
echo "Updated content" >> master.txt
git add master.txt
git commit -m "Updated master.txt in private"

**#### 7. Update public1 & public2 with private changes**

git checkout public1
git merge private

git checkout public2
git merge private

**#### 8. Update master with private**

git checkout master
git merge private

**#### 9. Update private with master**

git checkout private
git merge master

**Q 5. Tasks to Be Performed:**

1. Create a Git Flow workflow architecture on Git

2. Create all the required branches

3. starting from e feature branch, push the branch to the master, following the architecture

4. Push an urgent.txt on master using hotfix

**#### Initialize repository**

mkdir GitFlowAssignment
cd GitFlowAssignment
git init

**#### Create develop branch**

git checkout -b develop

**#### Create feature branch**

git checkout -b feature/login

touch login.txt
git add login.txt
git commit -m "Added login feature"

**#### Merge feature into develop**

git checkout develop
git merge feature/login

**#### Merge develop into master**

git checkout master
git merge develop

**#### Create hotfix branch**

git checkout -b hotfix/urgent

touch urgent.txt
git add urgent.txt
git commit -m "Added urgent hotfix"

**#### Merge hotfix into master**

git checkout master
git merge hotfix/urgent

**#### Merge hotfix back to develop**

git checkout develop
git merge hotfix/urgent
