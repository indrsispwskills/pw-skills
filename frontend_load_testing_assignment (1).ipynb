
## 2. Simple Load Testing Script

This script simulates **multiple users sending requests simultaneously**
to a web application.


````

import requests
import time
from concurrent.futures import ThreadPoolExecutor

URL = "https://jsonplaceholder.typicode.com/posts"

def send_request():
    start = time.time()
    response = requests.get(URL)
    end = time.time()

    return {
        "status": response.status_code,
        "time": end - start
    }

def run_load_test(users=20):
    results = []

    with ThreadPoolExecutor(max_workers=users) as executor:
        futures = [executor.submit(send_request) for _ in range(users)]

        for future in futures:
            results.append(future.result())

    return results
````


## 3. Run the Load Test


````

results = run_load_test(50)

times = [r["time"] for r in results]

print("Total Requests:", len(results))
print("Average Response Time:", sum(times)/len(times))
print("Fastest Response:", min(times))
print("Slowest Response:", max(times))
````


## 4. Analyze Results

Important metrics:

- **Average Response Time**
- **Maximum Response Time**
- **Error Rate**
- **Throughput**

If response time increases with more users, the application may have
a **performance bottleneck**.

Common bottlenecks include:

- Slow database queries
- Server CPU overload
- Network latency
- Inefficient frontend API calls
