```
Q  7: :Simulate a dataset from a binomial distribution (n = 10, p = 0.5) using
NumPy and plot the histogram.
```

````python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. Define binomial distribution parameters ---
n = 10  # Number of trials
p = 0.5 # Probability of success on each trial
size = 1000 # Number of times to simulate the binomial experiment

print(f"Simulating a Binomial Distribution with n={n}, p={p}, and size={size} observations.\n")

# --- 2. Simulate the dataset ---
dataset = np.random.binomial(n=n, p=p, size=size)

# Display the first few simulated values and some basic statistics
print(f"First 10 simulated values: {dataset[:10]}")
print(f"Mean of simulated data: {np.mean(dataset):.2f}")
print(f"Standard Deviation of simulated data: {np.std(dataset):.2f}\n")

# --- 3. Plot the histogram ---
plt.figure(figsize=(10, 6))
plt.hist(dataset, bins=np.arange(n + 2) - 0.5, edgecolor='black', alpha=0.7)
# np.arange(n + 2) - 0.5 creates bins centered on integer values from 0 to n

plt.title(f'Histogram of Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.xticks(np.arange(n + 1)) # Ensure x-axis ticks are at integer values
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
````
