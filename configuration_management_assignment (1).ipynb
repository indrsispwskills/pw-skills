
# Configuration Management Assignment – DevOps





## 1. Research and compare different configuration management tools (Ansible,Puppet, Chef) and write a report on their use cases and advantages?

Configuration management tools automate the setup, configuration, and maintenance of infrastructure.

They help DevOps teams ensure:

- Consistency across servers
- Automated deployment
- Faster environment setup



## Comparison: Ansible vs Puppet vs Chef

| Feature | Ansible | Puppet | Chef |
|-------|--------|--------|------|
| Architecture | Agentless | Agent-based | Agent-based |
| Language | YAML | Puppet DSL | Ruby |
| Learning Curve | Easy | Moderate | Hard |
| Communication | SSH | Puppet Agent | Chef Client |
| Setup Complexity | Simple | Moderate | Complex |
| Popular Use Case | Automation & provisioning | Enterprise infrastructure | Large-scale configuration |



## Ansible

### Advantages
- Agentless architecture
- Uses SSH
- Easy YAML playbooks
- Quick automation

### Use Cases
- Server provisioning
- Application deployment
- Infrastructure automation



## Puppet

### Advantages
- Strong enterprise support
- Scalable infrastructure management
- Declarative configuration

### Use Cases
- Large enterprise environments
- Compliance management



## Chef

### Advantages
- Powerful automation
- Infrastructure as Code
- Highly customizable

### Use Cases
- Large distributed systems
- Cloud infrastructure automation



## 2. Extend the CI pipeline to include continuous delivery using Jenkins or GitLab CI. Implement a blue-green deployment strategy?

Continuous Delivery ensures code changes are automatically built, tested, and prepared for deployment.

Common CI/CD tools:

- Jenkins
- GitLab CI
- GitHub Actions



## Jenkins Pipeline Example


````

pipeline {
    agent any

    stages {

        stage('Build') {
            steps {
                echo 'Building application...'
            }
        }

        stage('Test') {
            steps {
                echo 'Running tests...'
            }
        }

        stage('Deploy') {
            steps {
                echo 'Deploying application...'
            }
        }
    }
}
````


## GitLab CI Example


````

# .gitlab-ci.yml

stages:
  - build
  - test
  - deploy

build_job:
  stage: build
  script:
    - echo "Building project"

test_job:
  stage: test
  script:
    - echo "Running tests"

deploy_job:
  stage: deploy
  script:
    - echo "Deploying application"
````


## Blue-Green Deployment Strategy

Blue‑Green deployment reduces downtime during releases.

### How it works

```
Users
  |
Load Balancer
 /      Blue   Green
(Old)  (New)
```

Steps:

1. Deploy new version to **Green environment**
2. Test the Green environment
3. Switch traffic from Blue → Green
4. Keep Blue as backup



## Advantages of Blue‑Green Deployment

- Zero downtime deployment
- Easy rollback
- Safer production releases



## Conclusion

Configuration management tools like **Ansible, Puppet, and Chef** help automate infrastructure.

CI/CD pipelines with **Jenkins or GitLab CI** ensure faster and reliable deployments.

Using **Blue‑Green deployment** improves release safety and minimizes downtime.
