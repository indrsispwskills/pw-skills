### 1.  What is the difference between a function and a method in Python?

**Function**
- A block of reusable code defined using `def`.
- It is not associated with any object.

Example:
```python
def add(a, b):
    return a + b

print(add(5,3))
```

**Method**
- A function that belongs to a class/object.

Example:
```python
text = "hello"
print(text.upper())
```

`upper()` is a method of the string object.

### 6. Explain the concept of generators in Python and how they are defined

Generators produce values one at a time using `yield`.

Example:

```python
def count_up(n):
    for i in range(n):
        yield i

for num in count_up(5):
    print(num)
```



### 7. What are the advantages of using generators over regular functions?


- Memory efficient
- Produce values lazily
- Useful for large datasets

Example:
```python
def squares(n):
    for i in range(n):
        yield i*i
```

### 9. Explain the purpose and usage of the `map()` function in Python.

`map()` applies a function to every element of an iterable.

Example:

```python

result = list(map(lambda x: x*x, nums))
print(result)
```



### 10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

| Function | Purpose |
|--------|--------|
| map() | Apply function to each element |
| filter() | Filter elements based on condition |
| reduce() | Reduce iterable to single value |

Example:

```python
from functools import reduce

nums=[1,2,3,4]

print(list(map(lambda x:x*2, nums)))
print(list(filter(lambda x:x%2==0, nums)))
print(reduce(lambda a,b:a+b, nums))
```

````python

# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list

def sum_even(numbers):
    return sum(n for n in numbers if n%2==0)

print(sum_even([1,2,3,4,5,6]))
````

````python

# 4 Write a Python function that checks if a given number is prime or not from 1 to 200.



def is_prime(n):
    if n<2:
        return False
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            return False
    return True

primes=[n for n in range(1,201) if is_prime(n)]
print(primes)
````

````python

# 5 Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

class Fibonacci:
    def __init__(self, limit):
        self.limit=limit
        self.a=0
        self.b=1
        self.count=0

    def __iter__(self):
        return self

    def __next__(self):
        if self.count>=self.limit:
            raise StopIteration

        value=self.a
        self.a,self.b=self.b,self.a+self.b
        self.count+=1
        return value

for num in Fibonacci(10):
    print(num)
````

````python

# 6 Write a generator function in Python that yields the powers of 2 up to a given exponent.


def powers_of_two(n):
    for i in range(n+1):
        yield 2**i

for p in powers_of_two(5):
    print(p)
````

````python

# 7 Implement a generator function that reads a file line by line and yields each line as a string.

def read_file(file):
    with open(file) as f:
        for line in f:
            yield line.strip()

# demo file
with open("sample.txt","w") as f:
    f.write("Line1\nLine2\nLine3")

for line in read_file("sample.txt"):
    print(line)
````

````python

# 9 Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

celsius=[0,10,20,30]

fahrenheit=list(map(lambda c:(c*9/5)+32,celsius))

print(fahrenheit)
````

````python

# 10  Create a Python program that uses `filter()` to remove all the vowels from a given string

vowels='aeiou'

text="Hello Python World"

result=''.join(filter(lambda x:x.lower() not in vowels,text))

print(result)
````
