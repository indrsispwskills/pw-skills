## 1. Sample Application Code

```

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        raise ValueError("Division by zero not allowed")
    return a / b
```

## 2. Unit Tests (using unittest)

```

import unittest

class TestMathFunctions(unittest.TestCase):

    def test_add(self):
        self.assertEqual(add(2,3), 5)

    def test_subtract(self):
        self.assertEqual(subtract(5,2), 3)

    def test_multiply(self):
        self.assertEqual(multiply(4,3), 12)

    def test_divide(self):
        self.assertEqual(divide(10,2), 5)

    def test_divide_by_zero(self):
        with self.assertRaises(ValueError):
            divide(10,0)


unittest.main(argv=[''], verbosity=2, exit=False)
```


### Unit Test Explanation
Unit tests verify that **individual functions work correctly**.
Each function is tested independently.


## 3. Integration Testing Example

```

import requests

def test_api():
    url = "https://jsonplaceholder.typicode.com/posts/1"
    response = requests.get(url)

    assert response.status_code == 200

    data = response.json()
    print("API Title:", data["title"])

test_api()
```


### Integration Test Explanation
Integration testing checks how **multiple components interact**.
Example: Application communicating with an external API.


## 4. End‑to‑End Testing Example (Selenium)

```

# Example E2E structure using Selenium
# !pip install selenium

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys

def selenium_test():
    driver = webdriver.Chrome()

    driver.get("https://www.google.com")

    search_box = driver.find_element(By.NAME, "q")
    search_box.send_keys("Python testing tutorial")
    search_box.send_keys(Keys.RETURN)

    print("Page Title:", driver.title)

    driver.quit()

# selenium_test()   # Uncomment when Selenium driver is configured
```


### End‑to‑End Test Explanation
E2E tests simulate **real user behavior in a browser**, such as:

- Opening a website
- Entering search queries
- Clicking buttons
- Verifying results
