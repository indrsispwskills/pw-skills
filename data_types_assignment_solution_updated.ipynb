### 2. Explain the difference between mutable and immutable data types with examples

Mutable: Can be changed after creation.
Example: list, dictionary, set

Immutable: Cannot be changed after creation.
Example: string, tuple

Example:
```python
a = [1,2,3]
a.append(4)   # mutable

b = "hello"
# b[0] = 'H' -> error (immutable)
```

# 14. How does the “in” keyword work differently for lists and dictionaries?

The in keyword checks membership.

In Lists

It searches through each element sequentially.

numbers = [1,2,3,4]

print(3 in numbers)
In Dictionaries

It checks only keys, not values.

student = {"name":"John","age":21}

print("name" in student)

To check values:

print("John" in student.values())

# 16. What is a nested dictionary? Give an example.

A nested dictionary is a dictionary inside another dictionary.

It is useful for storing complex structured data.

Example

````python
students = {
    "student1": {"name":"Alice", "age":20},
    "student2": {"name":"Bob", "age":21}
}

print(students["student1"]["name"])
````

# 17. Describe the time complexity of accessing elements in a dictionary

Dictionary access has O(1) time complexity on average.

This means retrieving a value using a key is very fast, because dictionaries use hash tables internally.

Example

````python
data = {"a":1, "b":2, "c":3}

print(data["b"])
````

The value is accessed instantly.

# 19. Why are dictionaries considered unordered and how does that affect retrieval?

Traditionally, dictionaries were unordered, meaning elements were not stored in any specific order.

This does not affect retrieval because values are accessed using keys, not positions.

Example

Even if order changes, the correct value is returned.
