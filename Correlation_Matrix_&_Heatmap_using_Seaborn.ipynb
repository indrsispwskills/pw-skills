## **6. Selecting Numeric Columns & Calculating Correlation Matrix**

```python
# Select numeric columns only
num_df = df.select_dtypes(include=['int64', 'float64'])

# Calculate correlation matrix

corr_matrix
```

The correlation matrix shows numerical relationships between exam scores.

---

## **7. Visualizing the Correlation Matrix using Seaborn Heatmap**

```python
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    square=True
)
plt.title("Correlation Matrix Heatmap ‚Äì Students Performance Dataset")
plt.tight_layout()
plt.show()
```

---

## **8. Interpretation of Results**

From the heatmap:

* **Reading score** and **Writing score** have a **very high positive correlation** (~0.95).
* **Math score** also strongly correlates with **Reading** and **Writing** scores (~0.80).
* This indicates students who perform well in one subject tend to perform well in others.

No negative correlations exist among numeric variables in this dataset.

````python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import kagglehub
import os
````

````python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("spscientist/students-performance-in-exams")
print("Path:", path)
print("Files:", os.listdir(path))
````

````python
csv_path = os.path.join(path, "StudentsPerformance.csv")
df = pd.read_csv(csv_path)

df.head()
````

````python
# Select numeric columns only
num_df = df.select_dtypes(include=['int64', 'float64'])

# Calculate correlation matrix
corr_matrix = num_df.corr()
corr_matrix
````

````python
plt.figure(figsize=(8, 6))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    square=True
)
plt.title("Correlation Matrix Heatmap ‚Äì Students Performance Dataset")
plt.tight_layout()
plt.show()
````
