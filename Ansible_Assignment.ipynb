## 1. Install and configure Ansible, Puppet, or Chef on a local or cloud environment. Write and execute basic configuration scripts?


## 1. What is Ansible?

Ansible is a **DevOps automation tool** used for:

- Configuration management
- Application deployment
- Infrastructure automation

### Key Features

- Agentless architecture
- Uses SSH for communication
- Simple YAML playbooks
- Easy to learn



## 2. Install Ansible (Ubuntu/Linux)

Run the following commands:

```bash
sudo apt update
sudo apt install ansible -y
```

Verify installation:

```bash
ansible --version
```



## 3. Configure Ansible Inventory

Ansible inventory defines the servers that will be managed.

Example `inventory.ini` file:


````

# inventory.ini

[webservers]
localhost ansible_connection=local
````


## 4. Test Ansible Connection

Use the `ping` module to verify connectivity.


````

# ansible webservers -i inventory.ini -m ping
````


## 5. Basic Ansible Playbook

A playbook is written in **YAML format** and contains tasks to execute on managed hosts.


````

# basic_playbook.yml

---
- name: Basic Ansible Configuration Example
  hosts: webservers
  become: yes

  tasks:

  - name: Install Nginx
    apt:
      name: nginx
      state: present
      update_cache: yes

  - name: Start Nginx Service
    service:
      name: nginx
      state: started
      enabled: yes
````


## 6. Execute the Playbook

Run the following command:

```bash
ansible-playbook -i inventory.ini basic_playbook.yml
```



## 7. Verify the Setup

After execution:

- Nginx will be installed
- Service will start automatically

Open the browser:

```
http://localhost
```



## 8. Basic Puppet Example (Optional)

Example Puppet manifest:

```
package { 'nginx':
  ensure => installed,
}

service { 'nginx':
  ensure => running,
  enable => true,
}
```



## 9. Basic Chef Example (Optional)

Example Chef recipe:

```
package 'nginx' do
  action :install
end

service 'nginx' do
  action [:enable, :start]
end
```



## Conclusion

Using configuration management tools like:

- **Ansible**
- **Puppet**
- **Chef**

DevOps engineers can automate infrastructure setup, ensuring consistency and faster deployment.
