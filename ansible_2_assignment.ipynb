
# Assignment –  Ansible-2




## 1. Write Ansible playbooks to automate the setup and configuration of a web server (e.g., Apache or Nginx)?

**Ansible** is an open‑source DevOps automation tool used for:

- Configuration Management
- Application Deployment
- Infrastructure Provisioning
- Automation

### Key Features

- Agentless (no software required on target machines)
- Uses SSH
- YAML based playbooks
- Simple and human readable



## 2. Ansible Architecture

```
           +----------------------+
           |   Ansible Control    |
           |        Node          |
           +----------+-----------+
                      |
                      | SSH
                      |
     --------------------------------------
     |                |                   |
+---------+     +---------+         +---------+
| Server1 |     | Server2 |         | Server3 |
| Apache  |     | Apache  |         | Apache  |
+---------+     +---------+         +---------+
```

- **Control Node** → machine running Ansible
- **Managed Nodes** → servers configured using Ansible



## 3. Installing Ansible

Example installation on Ubuntu:

```bash
sudo apt update
sudo apt install ansible -y
ansible --version
```



## 4. Ansible Inventory File

Inventory defines the servers to manage.


````

# inventory.ini

[webservers]
192.168.1.10
192.168.1.11
192.168.1.12
````


## 5. Ansible Playbook Structure

Typical playbook structure:

```
- hosts
- tasks
- modules
```



## 6. Apache Web Server Playbook


````

# apache_setup.yml

---
- name: Install and Configure Apache Web Server
  hosts: webservers
  become: yes

  tasks:

  - name: Install Apache
    apt:
      name: apache2
      state: present
      update_cache: yes

  - name: Start Apache Service
    service:
      name: apache2
      state: started
      enabled: yes

  - name: Deploy Sample Web Page
    copy:
      dest: /var/www/html/index.html
      content: |
        <html>
        <h1>Web Server Configured using Ansible</h1>
        <p>DevOps Automation Successful</p>
        </html>
````


## 7. Run the Apache Playbook

```bash
ansible-playbook -i inventory.ini apache_setup.yml
```



## 8. Nginx Web Server Playbook


````

# nginx_setup.yml

---
- name: Install and Configure Nginx
  hosts: webservers
  become: yes

  tasks:

  - name: Install Nginx
    apt:
      name: nginx
      state: present
      update_cache: yes

  - name: Start Nginx Service
    service:
      name: nginx
      state: started
      enabled: yes

  - name: Create Web Page
    copy:
      dest: /var/www/html/index.html
      content: |
        <html>
        <h1>Nginx Installed using Ansible</h1>
        </html>
````


## 9. Run the Nginx Playbook

```bash
ansible-playbook -i inventory.ini nginx_setup.yml
```

- Faster infrastructure deployment
- Eliminates manual configuration errors
- Easy scaling across multiple servers
- Reusable playbooks
- Agentless architecture
