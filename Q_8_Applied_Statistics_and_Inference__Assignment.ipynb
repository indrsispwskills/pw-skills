# Task
Generate multiple samples from a non-normal distribution, calculate the mean of each sample, and plot the distribution of these sample means alongside the original distribution to demonstrate the Central Limit Theorem. Provide an interpretation of the results.

## Simulate_NonNormal_Distribution

### Subtask:
Generate a dataset from a non-normal distribution (e.g., binomial, exponential, or Poisson) to serve as the population.


**Reasoning**:
To begin, I will import the numpy library, set the scale parameter for the exponential distribution, and then generate a large dataset from this distribution as instructed.



````python
import numpy as np

# Set the scale parameter for the exponential distribution
scale = 2

# Generate a large number of random samples from an exponential distribution
np.random.seed(42) # for reproducibility
non_normal_data = np.random.exponential(scale, size=100000)

print("Generated a non-normal dataset from an exponential distribution with shape:", non_normal_data.shape)
````

## Draw_Multiple_Samples

### Subtask:
Draw a large number of samples (e.g., 1000 or 10000) of a specified size (e.g., 30) from the non-normal distribution created in the previous step.


**Reasoning**:
To draw multiple samples from the non-normal distribution, I will define the number of samples and the size of each sample, then use a loop with `np.random.choice` to extract the samples and store them in a list.



````python
num_samples = 10000
sample_size = 30

samples = []

for _ in range(num_samples):
    sample = np.random.choice(non_normal_data, size=sample_size, replace=False)
    samples.append(sample)

print(f"Generated {len(samples)} samples, each of size {len(samples[0])}.")
print("First sample:", samples[0][:5], "...")
````

## Calculate_Sample_Means

### Subtask:
Calculate the mean of each sample drawn in the previous step.

**Reasoning**:
To calculate the mean of each sample, I will iterate through the `samples` list and apply `np.mean()` to each sample, storing the results in a new list called `sample_means`.

**Reasoning**:
To calculate the mean of each sample, I will iterate through the `samples` list and apply `np.mean()` to each sample, storing the results in a new list called `sample_means`.

**Reasoning**:
To implement the reasoning, I will now write the Python code to iterate through the generated samples and compute the mean for each, storing these means in a new list.



````python
sample_means = [np.mean(sample) for sample in samples]

print(f"Calculated means for {len(sample_means)} samples.")
print("First 5 sample means:", sample_means[:5])
````

## Plot_Distributions

### Subtask:
Plot the distribution of the sample means and compare it to the original non-normal distribution. Add labels, titles, and legends to the plots for clarity.

**Reasoning**:
To visualize the Central Limit Theorem, I will use `matplotlib.pyplot` and `seaborn` to create histograms of both the original non-normal data and the distribution of sample means. This will allow for a clear comparison to observe how the sample means tend towards a normal distribution, even when the original population is not normal.

**Reasoning**:
To visualize the distributions as described in the reasoning, I will use `matplotlib.pyplot` and `seaborn` to create two histograms: one for the original non-normal data and another for the calculated sample means. This will demonstrate the Central Limit Theorem by showing the sample means' distribution approaching normality.



````python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))

# Plot the original non-normal distribution
plt.subplot(1, 2, 1) # 1 row, 2 columns, first plot
sns.histplot(non_normal_data, kde=True, color='skyblue', bins=50)
plt.title('Distribution of Original Non-Normal Data (Exponential)')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Plot the distribution of sample means
plt.subplot(1, 2, 2) # 1 row, 2 columns, second plot
sns.histplot(sample_means, kde=True, color='lightcoral', bins=30)
plt.title('Distribution of Sample Means')
plt.xlabel('Sample Mean')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

print("Plotted the original non-normal distribution and the distribution of sample means.")
````

## Interpret_Results

### Subtask:
Provide an interpretation of the results observed in the plots, specifically how they demonstrate the Central Limit Theorem.

**Interpretation of Results:**The plots clearly demonstrate the Central Limit Theorem (CLT). The first plot shows the original distribution of the data generated from an exponential distribution, which is distinctly non-normal (skewed to the right). In contrast, the second plot displays the distribution of the means of 10,000 samples, each of size 30, drawn from this non-normal population. This distribution of sample means, despite coming from a non-normal parent distribution, approximates a normal (bell-shaped) distribution.This transformation from a skewed original distribution to a roughly normal distribution of sample means is the core concept of the Central Limit Theorem. As the theorem states, regardless of the shape of the population distribution, the distribution of sample means will tend towards a normal distribution as the sample size increases. This holds true as long as the sample size is sufficiently large, which was 30 in this case. The narrower spread of the sample means distribution compared to the original data also indicates that sample means are less variable than individual observations from the population.
