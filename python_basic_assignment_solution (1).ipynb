
## 3.  Compare and contrast mutable and immutable objects in Python with examples

**Mutable Objects**
Objects that can be modified after creation.

Examples:
- list
- dictionary
- set

**Immutable Objects**
Objects that cannot be changed after creation.

Examples:
- string
- tuple
- int


```python

# Mutable example
lst = [1,2,3]
lst.append(4)
print(lst)

# Immutable example
text = "hello"
print(text.upper())
```


## 7. Describe the different types of loops in Python and their use cases with examples

Loops are used to repeat a block of code.

Types of loops:

**1. For Loop**
Used to iterate over sequences.

**2. While Loop**
Runs until a condition becomes false.


```python

# For loop example

for i in range(5):
    print(i)
```

```python

# While loop example

i = 0
while i < 5:
    print(i)
    i += 1
```
