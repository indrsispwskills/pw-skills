`
Write a Python program that logs an error message to a log file when a
division by zero exception occurs.
`

```python
import logging

# Configure logging to write to a file named 'error.log'
# The level is set to INFO, so INFO, WARNING, ERROR, and CRITICAL messages will be logged.
# The format includes the timestamp, log level, and the message.
logging.basicConfig(filename='error.log', level=logging.INFO,
                    format='%(asctime)s - %(levelname)s - %(message)s')

def divide_numbers(numerator, denominator):
    try:
        result = numerator / denominator
        logging.info(f"Division successful: {numerator} / {denominator} = {result}")
        return result
    except ZeroDivisionError as e:
        error_message = f"Error: Division by zero occurred when dividing {numerator} by {denominator}."
        logging.error(error_message, exc_info=True) # exc_info=True adds exception info to the log
        return None

# Example usage:
print("Attempting a valid division...")
divide_numbers(10, 2)

print("Attempting a division by zero...")
divide_numbers(5, 0)

print("Check 'error.log' file for logs.")
```
